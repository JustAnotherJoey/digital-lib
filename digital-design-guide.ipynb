# Digital Design Guide

## Overview and Summary
Digital design is the art of designing circuits based on digital logic. Digital logic is discrete, meaning that the information is quantized(non-continuous) and stored in a digital format.

## Boolean Algebra
Boolean algebra is a system of mathematics which can be used to represent and model logical systems with binary variables. A binary variable is a variable which can take one of two discrete values, and binary variables can be combined using operations such as AND, OR, NAND, NOR, XOR, , NOT. In order to describe some sort of logical system, we must enumerate the output for each possible input to the system. Here are tables, sometimes called K-maps or truth tables, which provide descriptions of each of the fundamental logic operations
